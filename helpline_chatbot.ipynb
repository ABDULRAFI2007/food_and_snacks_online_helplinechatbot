```python
# pip install langchain_google_genai

import os
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI

# -----------------------------
# Setup API Key
# -----------------------------
if not os.environ.get("GOOGLE_API_KEY"):
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter Gemini API Key: ")

# -----------------------------
# Allowed Food Delivery Keywords
# -----------------------------
FOOD_KEYWORDS = [
    "food", "snacks", "meal", "breakfast", "lunch", "dinner",
    "pizza", "burger", "sandwich", "fries", "biryani",
    "order", "buy", "price", "menu",
    "delivery", "track", "time",
    "cancel", "refund", "return",
    "payment", "cod", "upi", "card",
    "offer", "discount", "combo", "deal"
]

# -----------------------------
# Fixed Rejection Message
# -----------------------------
REJECT_MSG = "Sorry, I can only help with food and snacks delivery queries."

# -----------------------------
# Load Gemini Model
# -----------------------------
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

print("üçî Food & Snacks Delivery Bot (Type 'exit' to quit)\n")

# -----------------------------
# Chat Loop
# -----------------------------
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Goodbye üëã")
        break

    # ‚úÖ Manual Food Topic Filter
    if not any(keyword in user_input.lower() for keyword in FOOD_KEYWORDS):
        print("Bot:", REJECT_MSG)
        continue

    # ‚úÖ Only Food Queries Go to Gemini
    response = model.invoke(user_input)
    print("Bot:", response.content)
```
